# Histogram matching

```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.exposure import match_histograms

sns.set_style("whitegrid")
```

```python
worms_1 = cv2.imread('./images/worms.jpg', cv2.IMREAD_GRAYSCALE) # Converts image to grayscale
plt.imshow(worms_1, cmap='gray', vmin=0, vmax=255) # When using scalar data and no explicit norm, vmin and vmax define the data range that the colormap covers
plt.axis("off")
plt.show()
```

Adjusted image intensity

```python
worms_1_adjusted = worms_1 / worms_1.max() * 255 
fig, axs = plt.subplots(1, 2, figsize=(20, 5))
axs[0].imshow(worms_1_adjusted, cmap='gray', vmin=0, vmax=255) # When using scalar data and no explicit norm, vmin and vmax define the data range that the colormap covers
axs[0].axis("off")
axs[1].hist(worms_1_adjusted.ravel(), bins=255, range=(0, 255))
axs[1].set_ylim(0, 3e4)
plt.show()
```

```python
worms_2 = cv2.imread('./images/worms2.jpg', cv2.IMREAD_GRAYSCALE) # Converts image to grayscale
fig, axs = plt.subplots(1, 2, figsize=(20, 5))
axs[0].imshow(worms_2, cmap='gray', vmin=0, vmax=255) # When using scalar data and no explicit norm, vmin and vmax define the data range that the colormap covers
axs[0].axis("off")
axs[1].hist(worms_2.ravel(), bins=255, range=(0, 255))
plt.show()
```

Adjust an image I to match the intensity profile of a reference image Iref

```python
worms_2_adjusted = match_histograms(image=worms_2, reference=worms_1_adjusted)
fig, axs = plt.subplots(1, 2, figsize=(20, 5))
axs[0].imshow(worms_2_adjusted, cmap='gray', vmin=0, vmax=255) # When using scalar data and no explicit norm, vmin and vmax define the data range that the colormap covers
axs[0].axis("off")
axs[1].hist(worms_2_adjusted.ravel(), bins=255, range=(0, 255))
axs[1].set_ylim(0, 3e4)
plt.show()
```
