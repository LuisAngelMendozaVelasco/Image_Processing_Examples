# Increasing Contrast

```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
worms = cv2.imread('./images/worms.jpg', cv2.IMREAD_GRAYSCALE) # Converts image to grayscale
plt.imshow(worms, cmap='gray', vmin=0, vmax=255) # When using scalar data and no explicit norm, vmin and vmax define the data range that the colormap covers
plt.axis("off")
plt.show()
```

Histogram of an pixel intensities of an image.

```python
plt.hist(worms.ravel(), bins=255, range=(0, 255))
plt.ylim(0, 3e4)
plt.show()
```

Adjust intensity values to increase contrast. It adjusts the values to fill the entire available range by saturating the top and bottom 1% of intensity values.

```python
worms_adjusted = worms / worms.max() * 255
plt.imshow(worms_adjusted, cmap='gray', vmin=0, vmax=255) # When using scalar data and no explicit norm, vmin and vmax define the data range that the colormap covers
plt.axis("off")
plt.show()
```

```python
plt.hist(worms_adjusted.ravel(), bins=255, range=(0, 255))
plt.ylim(0, 3e4)
plt.show()
```
