# Rescale an Image by Blocks

```python
import cv2
import matplotlib.pyplot as plt
```

```python
book = cv2.imread("./images/book.jpg", cv2.IMREAD_GRAYSCALE)
book = book / book.max()
plt.imshow(book, cmap="gray")
plt.axis("off")
plt.show()
```

Adjusting the contrast block-by-block by scaling the intensities of a block to between 0 and 1.

```python
# Distinct block processing for image
def blockproc(image, block_size, function):
    h, w = image.shape
    m, n = block_size
    for x in range(0, h, m):
        for y in range(0, w, n):
            block = image[x:x+m, y:y+n]
            block[:, :] = function(block)

    return image

# Scales the input to the interval [0, 1]
def rescale(block):
    return block / block.max()
```

## 25x25 block size

```python
book_block = blockproc(book.copy(), (25, 25), rescale)
plt.imshow(book_block, cmap='gray')
plt.axis("off")
plt.show()
```

## 100x75 block size

```python
book_block = blockproc(book.copy(), (100, 100), rescale)
plt.imshow(book_block, cmap='gray')
plt.axis("off")
plt.show()
```
