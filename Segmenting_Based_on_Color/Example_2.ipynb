# Segmenting Ripe Strawberries by Using Both Red and Green Color Planes

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
strawberry = cv2.imread("./images/strawberryPlant.jpg")
strawberry = cv2.cvtColor(strawberry, cv2.COLOR_BGR2RGB)
red, green, blue = cv2.split(strawberry)

fig, axs = plt.subplots(1, 4, figsize=(20, 20))
axs[0].imshow(strawberry)
axs[0].axis('off')
axs[0].set_title('RGB')
axs[1].imshow(red, cmap='gray')
axs[1].axis('off')
axs[1].set_title('red plane')
axs[2].imshow(green, cmap='gray')
axs[2].axis('off')
axs[2].set_title('green plane')
axs[3].imshow(blue, cmap='gray')
axs[3].axis('off')
axs[3].set_title('blue plane')
plt.show()
```

```python
red_green_mask = (red > 175) & (green < 150)

plt.imshow(red_green_mask, cmap='gray')
plt.axis('off')
plt.show()
```

```python
difference_mask = (red - np.minimum(green, red)) > 50

plt.imshow(difference_mask, cmap='gray')
plt.axis('off')
plt.show()
```
