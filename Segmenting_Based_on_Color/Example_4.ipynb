# Segmenting Solar Flares

```python
import cv2
import matplotlib.pyplot as plt
```

```python
solar_flare = cv2.imread("./images/solarFlare.jpg")
solar_flare = cv2.cvtColor(solar_flare, cv2.COLOR_BGR2RGB)

plt.imshow(solar_flare)
plt.axis('off')
plt.show()
```

Converts the image from RGB to the HSV color space.

```python
solar_flare_HSV = cv2.cvtColor(solar_flare, cv2.COLOR_RGB2HSV) # https://en.wikipedia.org/wiki/HSL_and_HSV
```

Split the color spaces into their respective channels.

```python
hue, saturation, value = cv2.split(solar_flare_HSV)

fig, axs = plt.subplots(1, 3, figsize=(20, 20))
axs[0].imshow(hue, cmap='gray')
axs[0].set_title('Hue')
axs[0].axis('off')
axs[1].imshow(saturation, cmap='gray')
axs[1].set_title('Saturation')
axs[1].axis('off')
axs[2].imshow(value, cmap='gray')
axs[2].set_title('Value')
axs[2].axis('off')
plt.show()
```

Solar flares have low saturation and high value.

```python
solar_flare_mask = ((saturation / 255) < 0.5) & ((value / 255) > 0.95)

plt.imshow(solar_flare_mask, cmap='gray')
plt.axis('off')
plt.show()
```
