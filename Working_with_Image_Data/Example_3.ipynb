# Save Images to Image File

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
glacier_1980 = cv2.imread('./images/bearGlacier1980.png') # By default converts image to the 3 channel BGR color image
glacier_1980 = cv2.cvtColor(glacier_1980, cv2.COLOR_BGR2RGBA) # Converts an image from one color space to another

glacier_2011 = cv2.imread('./images/bearGlacier2011.png') # By default converts image to the 3 channel BGR color image
glacier_2011 = cv2.cvtColor(glacier_2011, cv2.COLOR_BGR2RGBA) # Converts an image from one color space to another

gs_1980 = cv2.cvtColor(glacier_1980, cv2.COLOR_RGBA2GRAY) # Converts to grayscale
gs_2011 = cv2.cvtColor(glacier_2011, cv2.COLOR_RGBA2GRAY) # Converts to grayscale
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Bear Glacier Comparison - 1980-201')
axs[0].imshow(gs_1980, cmap='gray')
axs[0].axis('off')
axs[1].imshow(gs_2011, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
cv2.imwrite("./images/gs_1980.png", gs_1980);
cv2.imwrite("./images/gs_2011.png", gs_2011);
```

Convert the images to colormaps with two colors

```python
glacier_1980 = Image.open("./images/bearGlacier1980.png")
ind_1980 = glacier_1980.quantize(colors=2) # Convert the image to 'P' mode with the specified number of colors
plt.imshow(ind_1980)
plt.axis('off')
plt.show()
```

```python
glacier_2011 = Image.open("./images/bearGlacier2011.png")
ind_2011 = glacier_2011.quantize(colors=2) # Convert the image to 'P' mode with the specified number of colors
plt.imshow(ind_2011)
plt.axis('off')
plt.show()
```

```python
ind_1980.save("./images/indGlacier_1980.png")
ind_2011.save("./images/indGlacier_2011.png")
```
