# Calculate the Difference Between the Images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
glacier_1980 = cv2.imread('./images/bearGlacier1980.png') # By default converts image to the 3 channel BGR color image
glacier_1980 = cv2.cvtColor(glacier_1980, cv2.COLOR_BGR2GRAY) # Converts an image from one color space to another
plt.imshow(glacier_1980, cmap='gray')
plt.axis("off")
plt.show()
```

```python
glacier_2011 = cv2.imread('./images/bearGlacier2011.png') # By default converts image to the 3 channel BGR color image
glacier_2011 = cv2.cvtColor(glacier_2011, cv2.COLOR_BGR2GRAY) # Converts an image from one color space to another
plt.imshow(glacier_2011, cmap='gray')
plt.axis("off")
plt.show()
```

```python
diff_ice = cv2.subtract(glacier_1980, glacier_2011)
fig, axs = plt.subplots(1, 3, figsize=(20, 20))
axs[0].imshow(glacier_1980, cmap='gray')
axs[0].set_title('Glacier 1980')
axs[0].axis("off")
axs[1].imshow(glacier_2011, cmap='gray')
axs[1].set_title('Glacier 2011')
axs[1].axis("off")
axs[2].imshow(diff_ice, cmap='gray')
axs[2].set_title('Glacier 1980 - Glacier 2011')
axs[2].axis("off")
plt.show()
```

Scale the image to values between 0 and 1

```python
glacier_1980 = glacier_1980 / 255
glacier_2011 = glacier_2011 / 255
diff_ice = cv2.subtract(glacier_1980, glacier_2011)
fig, axs = plt.subplots(1, 3, figsize=(20, 20))
axs[0].imshow(glacier_1980, cmap='gray')
axs[0].set_title('Glacier 1980')
axs[0].axis("off")
axs[1].imshow(glacier_2011, cmap='gray')
axs[1].set_title('Glacier 2011')
axs[1].axis("off")
axs[2].imshow(diff_ice, cmap='gray')
axs[2].set_title('Glacier 1980 - Glacier 2011')
axs[2].axis("off")
plt.show()
```

There was more ice in 2011 (dark areas) and more ice in 1980 (bright areas). The gray areas are where the images were about the same (both water, exposed land, or covered in ice)

Another way to create a difference visualization. Pixels where the first image is brighter are highlighted green. Pixels where the second input is brighter are highlighted magenta.

```python
diff_ice = np.dstack((glacier_2011, glacier_1980, glacier_2011))
plt.imshow(diff_ice)
plt.axis("off")
plt.show()
```
